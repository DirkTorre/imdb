# setup

```python
import requests
import gzip
import shutil
import os
import pandas as pd
import numpy as np
```

```python
base_url = "https://datasets.imdbws.com/"
title_basics_file = "title.basics.tsv"
title_basics_zip = "title.basics.tsv.gz"

BASE_URL = "https://datasets.imdbws.com/"
FILES = {
    "tit_bas_f": "title.basics.tsv",
    "tit_bas_z": "title.basics.tsv.gz"
    }
```

## Downloading files

```python
# Change to True is you want to update the imdb files
DOWNLOAD = True
```

```python
# remove if already excist
if DOWNLOAD:
    title_basics_file = os.path.join("data/",FILES["tit_bas_f"])
    title_basics_zip = os.path.join("data/",FILES["tit_bas_z"])
    title_basics_url = BASE_URL+FILES["tit_bas_z"]
    if not os.path.exists("data"):
        os.makedirs("data")
    if os.path.exists(title_basics_file):
        os.remove(title_basics_file)
    if os.path.exists(title_basics_zip):
        os.remove(title_basics_zip)
    
    # download files
    response = requests.get(title_basics_url)
    open(title_basics_zip , "wb").write(response.content)

    # unzip files
    with gzip.open(title_basics_zip, 'rb') as f_in:
        with open(title_basics_file, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
    
    # remove zips
    os.remove(title_basics_zip)
```

## wat maakt een film goed?
<b>Hiermee geven we de films een unbiased score.</b><br><br>
Hoe goed de samenwerking is, en hoe goed de personeelsleden zijn.
hoe goed de personeelsleden zijn komt van talent, maar vooral van ervaring!


<b>De ervaring voor iemand op een bepaald moment heeft kunnen we tellen.
Hoe vaak 2 personen hebben samengewerkt op een bepaald moment kunnen we ook tellen.</b> in het aantal films waar ze aan hebben gewerkt op dat moment (huidige film telt dus mee).

De grap is dat we nu de imdb score helemaal weg kunnen laten, en er geen bias meer is vanuit het publiek op wat dan ook.


<b>connecties die belangrijk zijn:</b>
- producer - geld: de producer moet niet een massa=kassa mentaliteit hebben, dit zorgt voor doorsnee films.
- director - producer: zorgt er voor dat er uberhoupt geld is
- director - acteur: zorgt er voor dat de acteurs de tekst goed overzetten
- director - writer: zorgt voor betere tekst
- director - cinematographer: zorgt voor mooie film
- director - editor: zorgt voor lekker lopende film (op scherm)
- director - composer: werken veel samen.

## wat maakt een film interesant?
<b>Hiermee halen we de bias van het publiek weg.</b><br><br>
Als de film beter is dan de gemiddelde mensen hem vinden.
Daarvoor kan ik ik mijn score systeem gebruiken en de bestaande imdb score.
Des te hoger mij score is vergelijken met de imdb score, des te interesanter de film.
Hiermee maak je bias van het publiek ongedaan! <b>Dit is de normaisatie factor!</b>

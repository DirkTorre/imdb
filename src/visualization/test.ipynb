```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.layouts import column

# Enable Bokeh output in Jupyter
output_notebook()

# Generate dummy data
num_samples = 100

df = pd.DataFrame({
    "watched": np.random.choice([True, False], num_samples),
    "priority": np.random.choice([True, False], num_samples),
    "netflix": np.random.choice([True, False], num_samples),
    "prime": np.random.choice([True, False], num_samples),
    "averageRating": np.random.uniform(1, 10, num_samples).astype(float),
    "numVotes": np.random.randint(0, 10000, num_samples).astype(int),
    "primaryTitle": [f"Movie {i}" for i in range(num_samples)],
    "originalTitle": [f"Original Movie {i}" for i in range(num_samples)],
    "startYear": np.random.randint(1950, 2025, num_samples).astype(int),
    "runtimeMinutes": np.random.randint(60, 180, num_samples).astype(int),
    "radius": np.random.randint(1, 5, num_samples).astype(int),
})

source = ColumnDataSource(df)


TOOLTIPS = [
    ("title", "@primaryTitle"),
    ("year", "@startYear")
]

# Create a Bokeh figure
fig = figure(title="Average Rating vs Number of Votes",
           x_axis_label="Number of Votes",
           y_axis_label="Average Rating",
           width=700, height=500, y_range=(0,10),
           tooltips=TOOLTIPS)



# Add scatter plot
fig.circle(x="numVotes", y="averageRating", name="temp", radius=50, source=df)

# fig.select(name="temp")

hover = fig.select_one(HoverTool)
hover.point_policy = "follow_mouse"
# And we reference the mayor in the tooltip
hover.tooltips = [
    ("title", "@primaryTitle"),
    ("year", "@startYear")
]



OPTIONS = ["foo", "bar", "baz", "quux"]

multi_choice = MultiChoice(value=["foo", "baz"], options=OPTIONS)
multi_choice.js_on_change("value", CustomJS(code="""
    console.log('multi_choice: value=' + this.value, this.toString())
"""))

# multi_choice.on_event() << https://docs.bokeh.org/en/latest/docs/user_guide/interaction/python_callbacks.html#ug-interaction-python-callbacks

# Show the plot
show(column(multi_choice, fig))
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, Switch, CustomJS
from bokeh.layouts import column

# Enable Bokeh output in Jupyter
output_notebook()

# Generate dummy data
num_samples = 100

df = pd.DataFrame({
    "priority": np.random.choice([True, False], num_samples),
    "averageRating": np.random.uniform(1, 10, num_samples).astype(float),
    "numVotes": np.random.randint(0, 10000, num_samples).astype(int),
})

# Create ColumnDataSource
source = ColumnDataSource(df)
filtered_source = ColumnDataSource(df[df["priority"]])

# Create a Bokeh figure
p = figure(title="Average Rating vs Number of Votes",
           x_axis_label="Number of Votes",
           y_axis_label="Average Rating",
           width=700, height=500)

p.circle("numVotes", "averageRating", source=source, size=10, color="blue", alpha=0.6)

# Create a switch widget
switch = Switch(active=False)

# CustomJS callback to toggle data source
callback = CustomJS(args=dict(source=source, filtered_source=filtered_source, switch=switch), code="""
    source.data = switch.active ? filtered_source.data : source.data;
    source.change.emit();
""")

switch.js_on_change("active", callback)

# Show layout with switch and plot
show(column(switch, p))
```
